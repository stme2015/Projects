```python
#import the modules
import pandas as pd
import geopandas as gpd
import urllib.request
import os
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
recycle=pd.read_csv('Recycling_Diversion_and_Capture_Rates.csv')
recycle.head()
```

```python
boroughABV = ['MN', 'BX', 'BK', 'QE', 'QW', 'SI']
borough = ['Manhattan', 'Bronx', 'Brooklyn', 'Queens', 'Staten Island']
boroughNum = [1, 2, 3, 4, 4, 5]
recycle['borough'] = ""
recycle['borough_cd'] = 1
recycle['borough_id'] = ""
```

```python
for i in range(len(recycle)):
    index = boroughABV.index(recycle.District[i][:2])
    recycle.borough[i] = boroughNum[index]
    recycle.borough_cd[i] = recycle.District[i][-2:]
    recycle.borough_id[i] = str(recycle['borough'][i]) + "__" + str(recycle['borough_cd'][i])
```

```python
gross = pd.read_csv('DSNY_Monthly_Tonnage_Data.csv')
gross.head()
```

```python
gross['YEAR'] = 1
gross['BOROUGH_INDEX'] = "1"
for i in range(len(gross)):
    gross['YEAR'][i] = gross['MONTH'][i][:4]
    month = gross['MONTH'][i][-2:]
    gross['MONTH'][i] = month
    gross['BOROUGH_INDEX'][i] = str(gross['BOROUGH_ID'][i]) + "__" + str(gross['COMMUNITYDISTRICT'][i])
gross.head()
```

```python
newcol = ['MONTH', 'YEAR', 'BOROUGH', 'BOROUGH_INDEX', 'REFUSETONSCOLLECTED', 'PAPERTONSCOLLECTED', 'MGPTONSCOLLECTED', 'RESORGANICSTONS', 'SCHOOLORGANICTONS', 'LEAVESORGANICTONS', 'XMASTREETONS', 'COMMUNITYDISTRICT', 'BOROUGH_ID']
gross = gross[newcol]
gross.head()
```

```python
population = pd.read_csv('New_York_City_Population_By_Community_Districts.csv')
population.head()
```

```python
borough = ['Manhattan', 'Bronx', 'Brooklyn', 'Queens', 'Staten Island']
boroughNum = [1, 2, 3, 4, 5]
population['Borough_num'] = 1
population['BOROUGH_INDEX'] = '1'

for i in range(len(population)):
    index = borough.index(population.Borough[i])
    population.Borough_num[i] = boroughNum[index]
    population.BOROUGH_INDEX[i] = str(population['Borough_num'][i]) + "__" + str(population['CD Number'][i])
population.head()
```

```python
pop = population[['2010 Population', 'BOROUGH_INDEX']]
pop.head()
```

```python
gross1 = pd.merge(gross, pop, on=['BOROUGH_INDEX'])
gross1.head()
```

```python
gross1['NORM_REFUSE'] = gross1['REFUSETONSCOLLECTED']/gross1['2010 Population']
gross1['NORM_PAPER'] = gross1['PAPERTONSCOLLECTED']/gross1['2010 Population']
gross1['NORM_MGP'] = gross1['MGPTONSCOLLECTED']/gross1['2010 Population']
gross1['NORM_RESORG'] = gross1['RESORGANICSTONS']/gross1['2010 Population']
gross1['NORM_SCHOOLORG'] = gross1['SCHOOLORGANICTONS']/gross1['2010 Population']
gross1['NORM_LEAVES'] = gross1['LEAVESORGANICTONS']/gross1['2010 Population']
gross1['NORM_XMAS'] = gross1['XMASTREETONS']/gross1['2010 Population']
gross1['NORM_DIVRATE'] = (gross1['NORM_PAPER']+gross1['NORM_MGP'])/(gross1['NORM_REFUSE']+gross1['NORM_PAPER']+gross1['NORM_MGP'])
gross1.head()
```
