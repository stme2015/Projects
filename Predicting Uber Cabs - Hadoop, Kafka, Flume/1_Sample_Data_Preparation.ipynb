```python
data = sc.textFile('C:/Users/HistoricalData/')
```

```python
data.count()
```

```python
fea_data = data.map(lambda line : line.split(','))
```

```python
sampleData = fea_data.sample(False, .02, 12345)
```

```python
def toCSVLine(data):
    return ','.join(str(d) for d in data)
```

```python
lines = sampleData.map(toCSVLine)
#13. Write the above RDD lines as text file
lines.coalesce(1).saveAsTextFile('file:///home/SampleData')
```
